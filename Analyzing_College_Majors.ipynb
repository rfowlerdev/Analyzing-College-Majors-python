## Analysis of Earnings Based on College Majors

### Main focus on Analysis by Visualizations

#### Import necessary packages

```python
import pandas as pd
import matplotlib as plt
%matplotlib inline
```

#### Read data into pandas dataframe

```python
recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.iloc[0]
```

```python
recent_grads.head()
```

```python
recent_grads.tail()
```

#### Drop rows with missing values to avoid issues with plotting

```python
raw_data_count = recent_grads.count()
print('Before dropping nulls:')
print(raw_data_count)
recent_grads = recent_grads.dropna()
print('After dropping nulls:')
cleaned_data_count = recent_grads.count()
print(cleaned_data_count)
```

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs. Sample_size')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', 
                  title='Unemployment_rate vs. Sample_size')
```
